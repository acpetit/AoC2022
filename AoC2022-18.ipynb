```python
droplet = pd.read_csv("input18.txt",header=None,names=["x","y","z"])
droplet +=1

drop3D = np.zeros(droplet.max()+2,dtype=int)
```

```python
for i,row in droplet.iterrows():
    drop3D[*(row)] = 1
```

```python
fig,ax =plt.subplots(4,5)
ax = ax.reshape(20)
for k in range(20):
    ax[k].imshow(drop3D[k+1])
    ax[k].axis("off")
```

```python
totarea = sum((np.diff(drop3D,axis=k)!=0).sum() for k in range(3))
totarea
```

```python
interior = sum(np.maximum.accumulate(drop3D,axis=k)+np.flip(np.maximum.accumulate(np.flip(drop3D,axis=k),axis=k),axis=k) for k in range(3))==6
sum((np.diff(interior,axis=k)!=0).sum() for k in range(3))
```
